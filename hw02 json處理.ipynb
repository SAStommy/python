```python
import json #引入json
test = open('C:/Users/hp8470/data/糧商資訊.json',encoding="utf8") #test中儲入json,並用utf8格式(中文不在utf8格式會出亂碼)
Data = json.load(test) #data中以運行jsom
type(Data) #找出資料類型
```

```python
Data.items()
```

```python
temp = Data.get('allcompany')
temp
```

```python
# 查詢永發的電話
ask="永發"
```

```python
for i in temp:
    if(ask==i.get('公司或商號名稱')):
       print(ask,'的電話是',i.get('糧商電話號碼'))
```

```python
# 找出糧倉界的百貨公司---會買稻米、小麥、麵粉、米食製品全部基礎糧食
count=0
for i in temp:
    if(i.get('經營糧食種類')=="稻米、小麥、麵粉、米食製品"):
        print(i.get('公司或商號名稱'),'---',i.get('縣市'))
        count=count+1
```

```python
print("全台有合計",count,"間百貨糧商")
```

```python
#按糧商登記證證號排序
allnumber = []
for i in temp:
    allnumber.append(i.get('糧商登記證證號'))
allnumber
```

```python
#用lambda進行排序判定,列舉每一個公司且把其對應糧商登記證證號放
allnumber_enumerate = enumerate(allnumber, start = 1)
allnumber_enumerate_sorted = sorted(allnumber_enumerate, key = lambda s: s[1], reverse = True)
allnumber_enumerate_sorted
```

```python
allnumber_enumerate_sorted[0][0]
```

```python
# 印出排序過後的糧商登記證證號與對應公司名字
for i in allnumber_enumerate_sorted:
    j = i[0] - 1
    print(temp[j].get('公司或商號名稱'), '是', temp[j].get('糧商登記證證號'))
```
