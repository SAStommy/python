```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import time
# 引入工具

headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}
# 模擬使用者打開web的身份

start=1 # 頁數
end=202 # 頁數
alltitles=[]
alldates=[]
alltimes=[]
alltemp=[]
alltemp1=[]
allcomments=[]
allviews=[]
allusers=[]
for i in range(start,end+1):
    url="https://forum.gamer.com.tw/B.php?page="+str(i)+"&bsn=29330" # 爬蟲網站---我爬巴哈的闇影詩章(遊戲)版
    html = requests.get(url, headers=headers) # 取得網站連線
    soup = BeautifulSoup(html.text, 'html.parser') # 把網站變成html格式
    titles = soup.find_all(attrs={"class":"b-list__main__title"}) #搜尋器
    comments = soup.find_all(attrs={"class":"b-list__count__number"}) #搜尋器
    datetimes = soup.find_all(attrs={"class":"b-list__time__edittime"}) #搜尋器
    writers = soup.find_all(attrs={"class":"b-list__time__user"}) #搜尋器
    #print(f"====================第{str(i)}頁====================")
    for title in titles:
        #print(title.text.strip())
        alltitles.append(title.text.strip()) #取得標題
    for comment in comments:
        #print(comment.text.strip().split("/\n"))
        alltemp.append(comment.text.strip().split("/\n")) #取得觀看數&留言
    for datetime in datetimes:
        #print(datetime.text.strip().split())
        alltemp1.append(datetime.text.strip().split()) #取得日期&時間
    for writer in writers:
        allusers.append(writer.text.strip()) #取得版主
i=0
while(i<len(alltemp)):
    allcomments.append(alltemp[i][0]) #取得觀看數
    allviews.append(alltemp[i][1]) #取得留言
    i+=1
i=0
while(i<len(alltemp1)):
    alldates.append(alltemp1[i][0]) #取得日期
    alltimes.append(alltemp1[i][1]) #取得時間
    i+=1
#把所需資料換成list


# 把資料放入dict,json的對應格式之一
```

```python
# 爬蟲內容
soup
```

```python
# 轉換成json
data={}
data["標題"]=alltitles
data["版主"]=allusers
data["日期"]=alldates
data["時間"]=alltimes
data["觀看數"]=allviews
data["留言"]=allcomments
#把資料放入dict
```

```python
datajson=json.dumps(data,ensure_ascii=False) #做好json格式
```

```python
print(datajson) #json成果
```

```python
# 轉換成csv
datacsv= pd.DataFrame(data) # 做好dataframe格式
datacsv.to_csv('sv.csv') # 產生csv檔
datacsv.to_json("sv.json") # 產生json檔
```
