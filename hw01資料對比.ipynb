```python
import pandas as pd # 引入panda---能讀、寫csv檔的工具
```

```python
# 讀並儲存csv檔,注:寫入完整路徑,因為"\"是跳脫字元,會造成python語法error,所以要把"\"改成"/"
information = pd.read_csv("C:/Users/hp8470/data/establish202101.csv") 
close = pd.read_csv("C:/Users/hp8470/data/close202208.csv")
```

```python
information.tail() # 查看表1
```

```python
close.tail() # 查看表2
```

```python
company1 = list(information['統一編號']) # 把表1的[統一編號]欄的data轉為list,準備進行[統一編號]欄比對
company2 = list(close['統一編號']) # 把表2的[統一編號]欄的data轉為list,準備進行[統一編號]欄比對
```

```python
set(company1).union(company2) # 聯集---意義:沒有
# 因為我的資料對比----4130項 vs 1983項---所以我先付上代碼,輸出結果可以慢慢向下找

# set(company1).intersection(company2)            交集---意義:找出那些在2021年1月成立的公司於2022年8月解散
# set(company1).difference(company2)              差集---意義:找出在2021年1月成立的公司並會在2022年8月後繼續營運的公司
# set(company1).symmetric_difference(company2)    對稱差集---意義:沒有
```

```python
set(company1).intersection(company2) # 交集---意義:找出那些在2021年1月成立的公司於2022年8月解散
```

```python
set(company1).difference(company2) # 差集---意義:找出在2021年1月成立的公司並會在2022年8月後繼續營運的公司
```

```python
set(company1).symmetric_difference(company2) # 對稱差集---意義:沒有
```
